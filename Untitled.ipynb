```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from datetime import datetime

BASE_URL = "https://www.banggood.com/Wholesale-USB-Hubs-ca-5006.html"
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/123.0 Safari/537.36"
}

def scrape_page(url):
    response = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(response.content, "html.parser")
    
    products = []
    product_divs = soup.find_all("div", class_="product-item")
    
    for item in product_divs:
        # Product name
        name_tag = item.find("a", class_="title")
        name = name_tag.get_text(strip=True) if name_tag else None
        
        # Product URL
        url = name_tag["href"] if name_tag and name_tag.has_attr("href") else None
        
        # Price
        price_tag = item.find("span", class_="price")
        price = price_tag.get_text(strip=True) if price_tag else None
        
        # Rating
        rating_tag = item.find("span", class_="star")
        rating = rating_tag.get_text(strip=True) if rating_tag else None
        
        # Reviews
        reviews_tag = item.find("span", class_="review")
        reviews = reviews_tag.get_text(strip=True) if reviews_tag else None
        
        products.append({
            "name": name,
            "price": price,
            "rating": rating,
            "reviews": reviews,
            "url": url
        })
    
    return products

def scrape_all_pages(base_url, max_pages=5):
    all_products = []
    page = 1
    
    while page <= max_pages:
        url = f"{base_url}?page={page}"
        print(f"Scraping page {page}...")
        products = scrape_page(url)
        
        if not products:
            break
        
        all_products.extend(products)
        page += 1
    
    return pd.DataFrame(all_products)

# Run scraper
df = scrape_all_pages(BASE_URL, max_pages=10)

# Save results
df.to_csv("banggood_usb_hubs.csv", index=False)
print(df.head())
```

```python
import pandas as pd

# Load data
df = pd.read_csv("Books.csv")

# Clean Price: "£51.77" → 51.77
df["Price"] = df["Price"].str.replace("£", "").astype(float)

# Map ratings from words → numbers
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}
df["Rating"] = df["Rating"].map(rating_map)

# Remove duplicates
df = df.drop_duplicates()

# Remove extra spaces in title
df["Title"] = df["Title"].str.strip()

# Save cleaned file
df.to_csv("Books_Cleaned.csv", index=False)

df.head()
```

```python
rating_counts = df["Rating"].value_counts()
print(rating_counts)
```

```python
avg_price_by_rating = df.groupby("Rating")["Price"].mean()
print(avg_price_by_rating)
```

```python
top_expensive = df.sort_values("Price", ascending=False).head(10)
print(top_expensive)
```

```python
import matplotlib.pyplot as plt

rating_counts = df["Rating"].value_counts().sort_index()

plt.figure(figsize=(7,5))
plt.bar(rating_counts.index, rating_counts.values)
plt.title("Number of Books by Rating")
plt.xlabel("Rating (1–5 Stars)")
plt.ylabel("Number of Books")
plt.show()
```

```python
avg_price = df.groupby("Rating")["Price"].mean()

plt.figure(figsize=(7,5))
plt.plot(avg_price.index, avg_price.values, marker="o")
plt.title("Average Book Price per Rating")
plt.xlabel("Rating")
plt.ylabel("Average Price (£)")
plt.grid(True)
plt.show()
```

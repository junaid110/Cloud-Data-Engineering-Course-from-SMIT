- Abstraction:
The concept of hiding the complex reality while exposing only the necessary parts. It helps to reduce programming complexity and effort.

```python
# abstract class
class shapes():
    
    #abstract method
    def sides(self):
        pass
```

```python
class square(shapes):

    # functionality here 
    def sides(self):
        print("Square has 4 sides")
```

```python
sq_obj = square()
sq_obj.sides()
```

```python
class circle(shapes):

    # functionality here
    def sides(self):
        print("Circle has no sides")
```

```python
ci_obj = circle()
ci_obj.sides()
```

### Ploymorphism
- The ability of different classes to respond to the same message (method call) in different ways. 
- This allows for code to work with objects of various classes as if they were objects of a common superclass.

```python
class Animal:
    def speak(self):
        return "Animal speaks"

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

# Create instances
dog = Dog()
cat = Cat()

# Call the speak method on different objects
print(dog.speak())  # Output: Woof!
print(cat.speak())  # Output: Meow!
```

## TASK2:

- use encapsulation for class i:e., `StudentLogin()`:
- initialize two static but private attributes i:e., `__user_name`, `__password`
- create one private methods i:e., name `__dbconectivity()` that will validate you username & password. if both are valid then should print **"valid user"** otherwise **"not valid"**
- create one public methods i:e., update password & it should update your private password

```python
class StudentLogin():

    # initializer
    def __init__(self):
        self.__user_name = "Arsalan"
        self.__password = "123456"

    # private method
    def __dbconnectivity(self, username, password):
        if (self.__user_name == username and self.__password == password):
            print('valid user')
        else:
            print('not valid')

        # public method: use to abstract db connection
    def login(self, username, password):
        self.__dbconnectivity(username, password)

        # public method
    def update_password(self, password):
        self.__password = password
        print('password update successfully!')
```

```python
student_obj = StudentLogin()
```

```python
student_obj.login("ali", "3458346")
```

```python
student_obj.login("Arsalan", "123456")
```

```python
student_obj.update_password("98765")
```

```python
student_obj.login("Arsalan", "123456")
```

```python
student_obj.login("Arsalan", "98765")
```

## Data files

- read  --> 'r'
- write --> 'w'
- append --> 'a'
- read/write --> 'w+'

```python
# METHOD1 - we ahve to specifically close the file
f = open('file.txt', 'w')
f.write("this is the contect of my file")
f.close()
```

```python
# METHOD2 - file will close automaticlally (Recommended Method)
with open('file.txt', 'w') as f:
    f.write('Now file will automatically close.') # this will overrite the content in file if file exists
```

```python
with open('file.txt', 'a') as f:
    # f.write('Now this line will get appended in the file. \nThis ill get appended on nect line')
    f.writelines("\nThis ia a new line")
```

```python
with open('file.txt', 'r') as f:
    file_content = f.read()
    print(file_content)
```

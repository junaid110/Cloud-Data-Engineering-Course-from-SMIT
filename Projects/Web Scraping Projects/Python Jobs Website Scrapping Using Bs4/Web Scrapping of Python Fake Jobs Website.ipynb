```python
from bs4 import BeautifulSoup
import requests 
```

```python
url = 'https://realpython.github.io/fake-jobs/'
request = requests.get(url)
request
```

```python
soup = BeautifulSoup(request.content,'html.parser')
soup
```

```python
complete_title = soup.find('div',class_ ='container mb-5') #tag, #xlass
print(complete_title.text.strip())
```

```python
title = soup.find('h1',class_ ='title is-1') #tag, #xlass
print(title.text.strip())
```

```python
job_title = soup.find('h2', class_= 'title is-5').text.strip()
job_sub_title = soup.find('h3', class_= 'subtitle is-6 company').text.strip()
location = soup.find('p', class_= 'location').text.strip()
date = soup.find('p', class_= 'is-small has-text-grey').text.strip()
print(job_title)
print(job_sub_title)
print(location)
print(date)
```

```python
#Find All Jobs

job_cards = soup.find_all('div', class_='column is-half')
print(len(job_cards))
print(job_cards)
```

```python
for i in job_cards:
    job_title = i.find('h2', class_= 'title is-5').text.strip()
    job_sub_title = i.find('h3', class_= 'subtitle is-6 company').text.strip()
    location = i.find('p', class_= 'location').text.strip()
    date = i.find('p', class_= 'is-small has-text-grey').text.strip()
    apply_link = i.find('a', string='Apply')
    jobs = [job_title,job_sub_title,location,date,apply_link]
    print(jobs)
```

```python
import pandas as pd

df = pd.DataFrame(jobs)
df.to_csv("jobs.csv", index=False)

print(" CSV exported using Pandas")
```
